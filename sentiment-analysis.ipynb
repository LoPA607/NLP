```python
import gensim
import pandas as pd
```

```python
df=pd.read_json("/kaggle/input/cell-phones-and-accessories-5-json/Cell_Phones_and_Accessories_5.json"
 , lines=True)
df
```

```python
#to remove unwantted letters and make upper case to lower case
review_text=df.reviewText.apply(gensim.utils.simple_preprocess)
```

```python
review_text
```

```python
model=gensim.models.Word2Vec(
        window=10,
        min_count=2,
        workers=4
)
```

```python
model.build_vocab(review_text, progress_per=1000)
```

```python
model.train(review_text, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.wv.similarity(w1="cheap" , w2="inexpensive")
```
